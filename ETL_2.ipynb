ESTE SEGUNDO PROCESO DE ETL SE LLEVA A CABO PARA PODER DEJAR SOLO LAS COLUMNAS UTILES QUE NECESITAN LAS FUNCIONES DE LOS ENDPOINTS. ESTE PROCESO DE OPTIMIZACION NOS ES UTIL PARA AGILIZAR EL DEPLOY EN RENDER.

```python
import pandas as pd
```

```python
credits_cast = pd.read_parquet('../PI01/credits_cast.parquet')
credits_crew = pd.read_parquet('../PI01/credits_crew.parquet')
movies = pd.read_parquet('../PI01/movies_con_etl.parquet')
```

```python
# Observamoso is hay valores repetidos para la columnas 'id_pelicula'.

movies['id_pelicula'].describe()
```

```python
movies['id_pelicula'].value_counts()
```

```python
movies_sin_duplicados = movies.drop_duplicates(subset=['id_pelicula'])
```

```python
movies_sin_duplicados.info()
```

```python
#Armamos un dataset con las columnas que son utiles para las primeras cuatro funciones que nos piden.

movies_funcion = movies_sin_duplicados[['id_pelicula','release_date','release_year','title','popularity','vote_count','vote_average']]
```

```python
movies_funcion.head()
```

```python
#Eliminamos los nulos.
movies_funcion.dropna(axis=0, inplace=True)
```

```python
#Agregamos columna 'month' para el endpoint 1.

movies_funcion['month'] = movies_funcion['release_date'].dt.month
```

```python
#Agregamos columna 'day' para el endpoint 2.

movies_funcion['day'] = movies_funcion['release_date'].dt.day_name()
```

```python
movies_funcion.head()
```

```python
movies_funcion.info()
```

```python
# Guardamos el archivo con el nombre 'movies_funcion'.

movies_funcion.to_parquet('movies_funcion.parquet', engine='pyarrow')
```
