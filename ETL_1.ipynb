PROCESO ETL PARA DATASET CREDITS Y MOVIES

```python
import pandas as pd
```

```python
# Hacemos la lectura de los datasets con formato csv.

credits = pd.read_csv('../PI01/credits.csv')
movies = pd.read_csv('../PI01/movies_dataset.csv')
```

```python
# Hacemos una vista inicial de los datos.

credits.head(3)
```

```python
movies.head(3)
```

```python
# Cambiamos el nombre de la columna 'id' por 'id_pelicula', para poder identificar este valor que es 
# importante para relacionar luego a credits con movies ademas de diferenciar los datos cuando sean desanidados.

credits.rename(columns={'id':'id_pelicula'},inplace=True)
credits
```

DATASET CREDITS

DADA LA CANTIDAD DE INFORMACION QUE TIENE ANIDADA EL DATASET 'CREDITS' SE DECIDE REALIZAR UNA DIVISION DE ESTE ARCHIVO A DOS DATASETS LLAMADOS 'CREDITS_CAST' Y 'CREDITS_CREW' 

```python
#LIMPIEZA DE CREDITS_CAST Y DESANIDADO.
```

```python
# Para el desanidado de los datos, vamos a usar la funcion explode(). 
# Esta nos permite separar los diccionarios en diferentes filas.
```

```python
credits_cast = credits[['id_pelicula','cast']]
```

```python
credits_cast.head()
```

```python
type(credits_cast.iloc[0,1])
```

```python
# Vemos que los datos de la columna 'cast' estan en formato string, tenemos que modificarlos. 
# Formula utilizada para convertir datos 'str' a listas de diccionarios. Esto me permite utilizar la funcion explode().

import ast

convertir_str_a_dict = lambda x: ast.literal_eval(x) if isinstance(x, str) else x
```

```python
credits_cast['cast'] = credits_cast['cast'].apply(convertir_str_a_dict)
type(credits_cast.iloc[0,1])
```

```python
# Una vez que esta en formato 'lista' podemos aplicar explode

credits_cast_exploded = credits_cast.explode('cast')
credits_cast_exploded
```

```python
# Una vez que tenemos los diccionarios desanidados, usamos la funcion 'json_normalize()' 
# para poder separar los datos en columnas.

credits_cast_expandido = pd.concat([credits_cast_exploded['id_pelicula'].reset_index(drop=True), pd.json_normalize(credits_cast_exploded['cast'].reset_index(drop=True))],axis=1)
credits_cast_expandido
```

```python
credits_cast_expandido.info()
```

```python
# Columnas que no me sirven: 'profile_path' (los datos no son de interes informativo ya que intentan mostrar una imagen)
```

```python
credits_cast_expandido.drop(columns='profile_path', inplace=True)
credits_cast_expandido.info()
```

```python
# Finalizado el proceso de desanidado, procedemos a guardar el archivo con el nombre 'credits_cast.parquet'. Se decide en 
# formato .parquet para hacerlo mas compacto.

credits_cast_expandido.to_parquet('credits_cast.parquet', engine='pyarrow')
```

```python
#LIMPIEZA DE CREDITS_CREW Y DESANIDADO.
```

```python
credits_crew = credits[['id_pelicula','crew']]
credits_crew
```

```python
type(credits_crew.iloc[0,1])
```

```python
# Se observa que es el mismo formato que el dataset anterior, por lo que se siguen los mismos pasos.
```

```python
credits_crew['crew'] = credits_crew['crew'].apply(convertir_str_a_dict)
type(credits_crew.iloc[0,1])
```

```python
credits_crew_exploded = credits_crew.explode('crew',ignore_index=True)
credits_crew_exploded
```

```python
credit_crew_expandido = pd.concat([credits_crew_exploded['id_pelicula'], pd.json_normalize(credits_crew_exploded['crew'])], axis=1)
credit_crew_expandido
```

```python
credit_crew_expandido.info()
```

```python
# La columna profile_path de antemano sabemos que no nos aporta informacion ademas de tener 
# muchos valores nulos, la eliminamos. 
```

```python
credit_crew_expandido.drop(columns='profile_path', inplace=True)
credit_crew_expandido.info()
```

```python
# Guardamos el dataset con el nombre 'credits_crew.parquet'.

credit_crew_expandido.to_parquet('credits_crew.parquet', engine='pyarrow')
```
